```python
import re

def txt_to_clean_string_array(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    clean_array = [re.sub(r"Human \d+:\s*", "", line.strip()) for line in lines if line.strip()]
    return clean_array

file_path = "./data/human_chat.txt"  # Metin dosyasƒ±nƒ±n adƒ±
clean_string_array = txt_to_clean_string_array(file_path)

print(clean_string_array)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

all_conversation_string = ' '.join(clean_string_array)
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(all_conversation_string)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud for Questions')
plt.axis('off')
plt.show()
```
